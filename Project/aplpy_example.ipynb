```python
import astropy
import matplotlib.pyplot as plt
import aplpy
import numpy as np
import warnings
import logging

logging.getLogger('astropy').setLevel(logging.WARNING) # suppress some of the extra outputs
warnings.filterwarnings('ignore') # suppress the extra warnings

co_cube = astropy.utils.data.get_pkg_data_filename('WFPC2u5780205r_c0fx.fits') # import data
fig = plt.figure(figsize = (10, 8)) # create the figure object
plots = [None] * 4 # create a list to store subplots

for i in range(len(plots)):
    plots[i] = aplpy.FITSFigure(co_cube, slices=[i], subplot = (2, 2, i + 1), figure = fig, dimensions = [0, 1]) # using aplpy to plot images 
    plots[i].show_colorscale() # show the color scale
    plots[i].add_grid() # add grid lines
    plots[i].tick_labels.set_font(size='x-small') # add ticks to axes
    plots[i].axis_labels.set_font(size='small') # adjust labels
    plots[i].add_colorbar() # add the colorbar
    plots[i].colorbar.set_axis_label_text('Pixels') # label the colorbar
    plots[i].set_title(f'CCD Image {i + 1}') # add subplot titles
    
fig.suptitle('CCD images of NGC4151') # add a figure title
plt.subplots_adjust(wspace=0.05) # adjust the spacing a little bit
plt.tight_layout() # tight layout fixes everything
plt.savefig('finalplot.png') # save it
```
